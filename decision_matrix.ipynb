```python
from openai import OpenAI

# OpenAI API Key
client = OpenAI(api_key="your_api")

# User Input
options = input("Enter options separated by commas: ")
criteria = input("Enter criteria separated by commas: ")

# Prompt
prompt = f"""
You are a Decision Matrix Builder.

Options:
{options}

Criteria:
{criteria}

Tasks:
1. Create a scoring table from 1-10
2. Assign reasonable weights to criteria
3. Calculate weighted totals
4. Recommend the best option
5. Explain why

Return output in clean text format.
"""

# OpenAI Response

response = client.responses.create(
    model="gpt-5",
    input=prompt
)

# Output

print("\n===== DECISION MATRIX =====\n")
print(response.output_text)
```

With UI integration

```python
# !pip install gradio openai

import gradio as gr
from openai import OpenAI

# OpenAI Client
client = OpenAI(api_key="YOUR_API_KEY")

# CORE FUNCTION
def decision_matrix(options, criteria):
    """
    Takes user input and returns AI-generated decision matrix.
    """

    prompt = f"""
You are a Decision Matrix Builder AI.

Options:
{options}

Criteria:
{criteria}

Tasks:
1. Assign weights (1-5) to each criterion
2. Score each option (1-10)
3. Build a decision matrix
4. Compute final weighted scores
5. Recommend the best option
6. Give a short explanation

Return in clean structured format.
"""

    response = client.responses.create(
        model="gpt-5",
        input=prompt
    )

    return response.output_text


# GRADIO UI
demo = gr.Interface(
    fn=decision_matrix,
    inputs=[
        gr.Textbox(label="Enter Options (comma-separated)"),
        gr.Textbox(label="Enter Criteria (comma-separated)")
    ],
    outputs=gr.Textbox(label="Decision Matrix Result"),
    title="🧠 Decision Matrix Builder",
    description="Compare multiple options using AI-based scoring and recommendation."
)


# RUN APP
demo.launch()
```
